```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import joblib
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from kmodes.kmodes import KModes
```

```python
df = pd.read_csv('cleaned_census.csv')
df.head(1)
```

```python
df = df.drop(['Unnamed: 0','nchild', 'nchild_under_5','nsibs',
             'eldch','sex', 'age'], axis=1)
```

```python
df.head()
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
df['ed_group_h'].unique()
```

```python
df['ed_group_h'] = np.where((df['ed_group_h'] == 1),
                           'Group 1', df['ed_group_h'])
df['ed_group_h'][0]
```

```python
df['ed_group_h'] = np.where((df['ed_group_h'] == '2'),
                           'Group 2', df['ed_group_h'])
df['ed_group_h'][1]
```

```python
df['ed_group_h'] = np.where((df['ed_group_h'] == '3'),
                           'Group 3', df['ed_group_h'])
df['ed_group_h'][3]
```

```python
df['ed_group_h'] = np.where((df['ed_group_h'] == '4'),
                           'Group 4', df['ed_group_h'])
df['ed_group_h'][2]
```

```python
df['ed_group_h'].unique()
```

```python
x = df.drop(['ed_group_h'], axis=1)
y = df['ed_group_h']
```

```python
kmodes_cao = KModes(n_clusters=4, init='Cao', verbose=1)
kmodes_cao.fit_predict(x)
```

```python
clusters = kmodes_cao.labels_
clusters = np.asanyarray(clusters)
```

```python
d = {'y': y,
     'clusters': clusters}
df = pd.DataFrame(data=d)
```

```python
df
```
